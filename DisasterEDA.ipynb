```python
import pandas as pd
import numpy as np
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

import nltk

```

```python
df = pd.read_csv('test.csv')
df.head()
```

```python

max(df.text, key=len)
```

```python
words = ' '.join(df['text']).split()
print(words)
len(set(words))
```

```python
words = [word for word in words if len(word) >=8]

word_counts = Counter(words)

# Найдите 10 наиболее встречающихся слов
top_10_words = word_counts.most_common(10)

# Выведите результат
print("10 наиболее встречающихся слов:")
for word, count in top_10_words:
    print(f"{word}: {count}")

len(words)
```

```python


# Преобразуйте отфильтрованные слова в строку
words1 = ' '.join(words)

# Создайте WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(words1)

# Отобразите WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
